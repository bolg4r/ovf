Провести интерполяционный полином Pn(x) через точки
$$x_k = 1 + \frac{k}{n},\; \; y_k = ln(x_k),\; \; k = 0, . . . , n$$
при n = 4, . . . , 15. Нарисовать графики Pn(x) − ln(x).

```python
from math import *
import numpy as np
from matplotlib import pyplot as plt
```

```python
def X(k):
    return 1 + k/N

def Y(k):
    return log1p(X(k))
```

![title](langrange_base_polynom.svg)

```python
def l(i, x):
    l = 1.
    for k in range(0, N+1):
        if k != i:
            l *= (x - X(k))
    return l
```

```python
def Newton(x):
    a = []
    for i in range(0, N+1):
        a.append(Y(i))
        for j in range(0, i):
            a[i] = (a[i] - a[j]) / (X(i) - X(j))
    res = a[N]
    for i in range(N-1, -1, -1):
        res = res * (x - X(i)) + a[i]
    return res
```

```python
fig, axs = plt.subplots(50, 1, figsize=(80, 120))
for N in range(4, 45):
    x_vals = []
    y_vals = []
    L_vals = []
    N_vals = []
    for i in range(0, N + 1):
        x = X(i)
        y = log1p(x)
        x_vals.append(x)
        y_vals.append(y)
        N_vals.append(Newton(x) - y)
    axs[N-4].plot(x_vals, N_vals, label = 'Ньютон')
    axs[N-4].set_title("График погрешности для полинома с N = {}".format(N))
    axs[N-4].set_xlabel('X')
    axs[N-4].set_ylabel('Погрешность')
    axs[N-4].legend(loc="upper right")

```
